```python
# Import pip package commands
# pip install serial
# pip install --upgrade google-cloud-texttospeech
# pip install playsound
# pip install pyttsx3

# Import Python Libraries
import serial
import pyttsx3
from playsound import playsound
import time
```

```python
# Program Variables

# Set up the serial port for the voice controller
# voice_controller = serial.Serial('COM4', 9600)
object_controller = serial.Serial('COM5', 9600)

# Set up the Text-to-Speech engine
engine = pyttsx3.init()

# Initialize Variables
getObjectsCondition = False
object_list = []

try:
    # Read voice_data from the serial port
    while True:
        object_data = object_controller.readline().decode('utf-8').strip()
        print("Data: " + object_data)

        # Change the condition variable to true if a prediction is starting
        if (object_data.startswith("Pred")):
            print("Condition: True")
            getObjectsCondition = True
            continue

        # Change the condition to false if a prediction is ending
        # If it is ending, print out the list of objects
        # Also, reset the list as well
        elif (object_data.startswith("Start") and getObjectsCondition == True):
            objects_str = ' '.join(map(str, object_list))
            print("The objects are " + objects_str)
            engine.say("The objects are " + objects_str)
            engine.runAndWait()
            object_list = []
            print("Condition: False")
            getObjectsCondition = False
            continue

        # If the condition to get objects is true but there's no objects,
        # print that to the console and set the condition to false
        if (object_data.startswith("No") and getObjectsCondition == True):
            print("Sorry, no objects were found.")
            engine.say("Sorry, no objects were found.")
            engine.runAndWait()
            print("Condition: False")
            getObjectsCondition = False

        # If the condition to get objects is true and there's an actual object, add it to the list
        if (getObjectsCondition == True and object_data):
            first_word = object_data.split()[0]
            print("Append: " + first_word)
            object_list.append(first_word)

# Catch the KeyboardInterrupt exception and close the serial port
except KeyboardInterrupt:
    print("Keyboard interrupt detected. Closing serial port.")
    # voice_controller.close()
    object_controller.close()
```
